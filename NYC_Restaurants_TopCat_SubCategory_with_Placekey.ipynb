Now we Fiter the 'city' column: city = 'New York City' in the each dataframe

```python
DF_core_poi_part1_NYC = DF_core_poi_part1.loc[DF_core_poi_part1['city'] == 'New York']
DF_core_poi_part2_NYC = DF_core_poi_part2.loc[DF_core_poi_part2['city'] == 'New York']
DF_core_poi_part3_NYC = DF_core_poi_part3.loc[DF_core_poi_part3['city'] == 'New York']
DF_core_poi_part4_NYC = DF_core_poi_part4.loc[DF_core_poi_part4['city'] == 'New York']
DF_core_poi_part5_NYC = DF_core_poi_part5.loc[DF_core_poi_part5['city'] == 'New York']

```

```python
DF_core_poi_part1_restaurants = DF_core_poi_part1_NYC.loc[DF_core_poi_part1_NYC['top_category'] == 'Restaurants and Other Eating Places']
DF_core_poi_part2_restaurants = DF_core_poi_part2_NYC.loc[DF_core_poi_part2_NYC['top_category'] == 'Restaurants and Other Eating Places']
DF_core_poi_part3_restaurants = DF_core_poi_part3_NYC.loc[DF_core_poi_part3_NYC['top_category'] == 'Restaurants and Other Eating Places']
DF_core_poi_part4_restaurants = DF_core_poi_part4_NYC.loc[DF_core_poi_part4_NYC['top_category'] == 'Restaurants and Other Eating Places']
DF_core_poi_part5_restaurants = DF_core_poi_part5_NYC.loc[DF_core_poi_part5_NYC['top_category'] == 'Restaurants and Other Eating Places']
```
