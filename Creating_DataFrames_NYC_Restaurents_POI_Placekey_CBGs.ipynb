```python
Dataframe_2_NYC = Dataframe_2.loc[Dataframe_2['city'] == 'New York']
```

Now we filter the 'top_category'in Dataframe_2_NY's to have only the 'Restaurants and Other Eating Places'

We would like to check the freequcnty count in the 'top_category' column to see the number of different business types such as "Restaurants and Other Eating Places" , "Personal Care Services, " Clothing Stores" etc... that are listed under 'top_category'.  
we can use the `.value_count()` to get the freequncy counts in the 'top_category'

```python
Dataframe_2_NYC['top_category'].value_counts()
```

Now we filter 'top_category' == 'Restaurants and Other Eating Places' to have only the restaurents in the New York City. (we  keep the 'sub_category' column in the dataframe for later analysis).

```python
Dataframe_2_NYC_restaurants = Dataframe_2_NYC.loc[Dataframe_2_NYC['top_category'] == 'Restaurants and Other Eating Places']
```

```python
Dataframe_2_NYC_restaurants.head()
```

Checking Dataframe_2_NY shape



Then I filtered the Dataframe_2_NYC_restaurants by  'placekey','top_category', 'sub_category','city','region','street_address' and create the dataframe called “df_placekey_topCatSubCat_nyc”

```python
df_placekey_topCatSubCat_nyc = Dataframe_2_NYC_restaurants[['placekey','top_category', 'sub_category','city','region','street_address']]
```

We create a copy of the 'df_placekey_topCatSubCat_nyc' dataframe and create a new dataframe called " Dataframe_3" 
( we do this step to shorten the name of the dataset and it not recommended re-copying dataframe if it is really big, sicne this is a guideline/documentation notebook for you to understand the process we did this step.

```python
df_placekey_topCatSubCat_nyc.shape
```

```python
Dataframe_3 = df_placekey_topCatSubCat_nyc
```

```python
Dataframe_3.head()
```

```python
Dataframe_3.shape
```

 Finally, I merged the Dataframe_1 and Dataframe_3  using the “placekey” and created the 4th dataframe (Dataframe_4)
 We write the Dataframe_4 to a CSV file named
 “nyc_poi_placekey_cbg_Restaurants_VisitCount_2018_01_01”


```python
Dataframe_4 = Dataframe_1.merge(Dataframe_3, on = 'placekey')
```
